```python
df = pd.read_csv("/content/import/UK-Export.csv")

df = df.drop(columns=["count","refused"], axis=0)
df.head(5)
```

```python
df_1 = df.copy()
sum_export_by_country = df_1.groupby('country')['value'].sum()
sum_export_by_country = sum_export_by_country.reset_index()
sum_export_by_country = sum_export_by_country.sort_values("value", ascending=False)
sum_export_by_country
```

```python
df_2 = df.copy()
sum_export_by_dates = df_2.groupby('date')['value'].sum()
sum_export_by_dates = sum_export_by_dates.reset_index()
#sum_export_by_dates = sum_export_by_dates.sort_values("value", ascending=False)
sum_export_by_dates.head(5)
```

```python
plt.figure(figsize=(16,6))
sns.lineplot(sum_export_by_dates, x=sum_export_by_dates['date'], y=sum_export_by_dates['value'])
```
